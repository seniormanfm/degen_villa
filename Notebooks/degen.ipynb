```python
# Example: Fetch and analyze Solana token transfer history for analytics
# Dependencies: $ python -m pip install requests python-dotenv pandas

import os
import requests
import json
import pandas as pd
from dotenv import load_dotenv

# Load API key from .env file
load_dotenv()
API_KEY = os.getenv("MORALIS_API_KEY")
if not API_KEY:
    raise ValueError("MORALIS_API_KEY not set in .env")

# Prompt user for contract address
contract_address = input("Enter the token contract address: ").strip()

# Fetch transfer history (Moralis Solana endpoint)
url = f"https://solana-gateway.moralis.io/token/mainnet/{contract_address}/transfers?limit=1000"
headers = {
    "Accept": "application/json",
    "X-API-Key": API_KEY
}
response = requests.get(url, headers=headers)
try:
    data = response.json()
except Exception as e:
    print("Error parsing response as JSON:", e)
    data = {}

# Convert to DataFrame for analysis
transfers = data.get("result", []) or data.get("transfers", [])
df = pd.DataFrame(transfers)

if df.empty:
    print("No transfer data found.")
else:
    # 1. Daily inflows and outflows
    df['blockTime'] = pd.to_datetime(df['blockTime'], unit='s')
    df['date'] = df['blockTime'].dt.date
    inflow = df.groupby('date')['amount'].sum()
    print("\nDaily total transfer volume:")
    print(inflow)

    # 2. Early buyers (first 10 unique senders)
    early_buyers = df.sort_values('blockTime').drop_duplicates('from', keep='first').head(10)
    print("\nEarly buyers (first 10 unique senders):")
    print(early_buyers[['from', 'blockTime', 'amount']])

    # 3. Amount bought per address (top 10)
    buyers = df.groupby('from')['amount'].sum().sort_values(ascending=False).head(10)
    print("\nTop 10 buyers by total amount:")
    print(buyers)

    # Note: To calculate profit, you need price at buy time and current price (requires price API)
    print("\nTo calculate profit, fetch historical and current price data for each buy.")
```

```python
# Calculate profit for top buyers (requires price API)
# Dependencies: $ python -m pip install requests python-dotenv pandas

import requests
import pandas as pd

# Assume df is already created from previous cell
# You need a price API for historical and current prices. We'll use Birdeye (Solana DEX aggregator) as an example.

# Prompt for token mint address (contract address)
token_address = input("Enter the token contract address again for price lookup: ").strip()

# Get current price from Birdeye API
price_url = f"https://public-api.birdeye.so/public/price?address={token_address}"
price_headers = {"X-API-KEY": "public"}  # Replace with your Birdeye API key if you have one
price_resp = requests.get(price_url, headers=price_headers)
current_price = None
try:
    price_data = price_resp.json()
    current_price = float(price_data.get('data', {}).get('value', 0))
except Exception as e:
    print("Error fetching current price:", e)

if current_price:
    print(f"Current price: {current_price} USD")
    # For each top buyer, estimate profit (assume all buys at average price, for demo purposes)
    # In reality, you need historical price at each buy timestamp
    top_buyers = df.groupby('from')['amount'].sum().sort_values(ascending=False).head(10)
    profit_results = []
    for buyer, total_bought in top_buyers.items():
        # Placeholder: assume average buy price is 80% of current price (for demo only)
        avg_buy_price = current_price * 0.8
        cost = total_bought * avg_buy_price
        current_value = total_bought * current_price
        profit = current_value - cost
        profit_pct = (profit / cost) * 100 if cost else 0
        profit_results.append({
            'buyer': buyer,
            'amount_bought': total_bought,
            'avg_buy_price': avg_buy_price,
            'current_value': current_value,
            'profit': profit,
            'profit_pct': profit_pct
        })
    profit_df = pd.DataFrame(profit_results)
    print("\nTop 10 buyers with estimated profit (demo):")
    print(profit_df)
else:
    print("Could not fetch current price. Profit calculation skipped.")
```
